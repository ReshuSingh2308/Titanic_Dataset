```python
sns.histplot(df['Age'],bins=20,kde=True)
plt.title('Age Distribution')

    
```

```python
sns.countplot(x='Sex', data=df)
plt.title('Passenger Gender Distribution')
plt.show()
```

```python
sns.scatterplot(x='Age', y='Fare', hue='2urvived', data=df)
plt.title('Age vs. Fare (Colored by Survival)')
plt.show()
```

```python
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title('Age Distribution by Passenger Class')
plt.show()
```

```python
survival_by_class = pd.crosstab(df['Pclass'], df['2urvived'])
sns.heatmap(survival_by_class, annot=True, fmt='d')
plt.title('Survival Count by Passenger Class')
plt.show()
```

```python
sns.pairplot(df[['Age', 'Fare', 'Pclass', '2urvived']], hue='2urvived')
plt.show()
```

```python
sns.boxplot(x=df['Fare'])
plt.title('Fare Distribution (Outlier Check)')
plt.show()
```

```python
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['Fare'] >= Q1 - 1.5*IQR) & (df['Fare'] <= Q3 + 1.5*IQR)]  # Remove outliers
```

```python
corr = df[['Age', 'Fare', 'Pclass', 'sibsp', 'Parch', '2urvived']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
